# Univariate Data Analysis

```python
# load packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

#suppress warnings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set(color_codes=True)

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

#load data
train = pd.read_pickle('../data/interim/train2_vrble_type_missing_value')
test = pd.read_pickle('../data/interim/test2_vrble_type_missing_value')
```

```python
# Visualize Distribution of continuous data
train_conti = []
for col in list(train.columns.values):
    if(train[col].dtypes == 'float64'):
        train_conti.append(col)

n_col =2
n_row = 10
for i in range(n_row):
    fg,ax = plt.subplots(nrows=1,ncols=n_col,figsize=(9,5))
    for j in range(n_col):
        sns.distplot(train.ix[:,train_conti].iloc[:, (i*n_col + j)],ax=ax[j])
```

```python
n_col =2
n_row = 10
for i in range(n_row):
    fg,ax = plt.subplots(nrows=1,ncols=n_col,figsize=(9,5))
    for j in range(n_col):
        sns.violinplot(train.ix[:,train_conti].iloc[:, (i*n_col + j)],ax=ax[j])
```
